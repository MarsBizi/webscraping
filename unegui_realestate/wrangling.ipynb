```python
# Clean data
#df3.flr_type.unique()
df3 = (pd.concat([df2, df], 1)
       .drop(['details_list'],1)
       .drop_duplicates()
       .query("district == ['–•–∞–Ω-–£—É–ª', '–°–æ–Ω–≥–∏–Ω–æ –•–∞–π—Ä—Ö–∞–Ω', '–°“Ø—Ö–±–∞–∞—Ç–∞—Ä', '–ë–∞—è–Ω–≥–æ–ª', '–ë–∞—è–Ω–∑“Ø—Ä—Ö','–ß–∏–Ω–≥—ç–ª—Ç—ç–π']")
       .reset_index(drop=True)
      )
df3.head()
df3 = (df3
       .replace({'flr_type':{'–ü–∞—Ä–∫–µ—Ç': 'parquet', '–ú–æ–¥':'wooden', '–õ–∞–º–∏–Ω–∞—Ç':'laminate', '–¶–µ–º–µ–Ω—Ç': 'tile', '–ß—É–ª—É—É':'tile', '–ü–ª–∏—Ç–∞': 'tile'}
                 , 'garage':{'–ë–∞–π—Ö–≥“Ø–π':0, '–ë–∞–π–≥–∞–∞':1}
                 , 'window_type':{'–í–∞–∫—É–º':'vinyl', '–¢”©–º”©—Ä –≤–∞–∫—É–º':'alum', '–ú–æ–¥':'wooden', '–ú–æ–¥–æ–Ω –≤–∞–∫—É–º':'clad_wood'}
                 , 'door_type':{'–ë“Ø—Ä–≥—ç–¥':'burged', '–¢”©–º”©—Ä –≤–∞–∫—É–º':'alum', '–¢”©–º”©—Ä':'metal', '–í–∞–∫—É–º':'vinyl', '–ú–æ–¥':'wooden'}
                 , 'mortgage':{'–õ–∏–∑–∏–Ω–≥–≥“Ø–π':'no', '–ë–∞–Ω–∫–Ω—ã –ª–∏–∑–∏–Ω–≥—Ç—ç–π':'bank', '–•—É–≤—å –ª–∏–∑–∏–Ω–≥—Ç—ç–π':'personal'}
                 , 'district':{'–•–∞–Ω-–£—É–ª':'KU', '–°–æ–Ω–≥–∏–Ω–æ –•–∞–π—Ä—Ö–∞–Ω':'SH', '–°“Ø—Ö–±–∞–∞—Ç–∞—Ä':'SU', '–ë–∞—è–Ω–≥–æ–ª':'BG', '–ë–∞—è–Ω–∑“Ø—Ä—Ö':'BZ', '–ß–∏–Ω–≥—ç–ª—Ç—ç–π':'CH'}
                })
       .assign(balcony = lambda x: x.balcony.str.extract('(\d+)')
               , area = lambda x: x.area.str.split(' ',1).str[0]
               , price = lambda x: x.price / 1000000
              )
       .replace({'balcony':{np.nan: 0}})
       .astype({'balcony':'category'
                , 'yr_built':'int'
                , 'building_height':'int'
                , 'area':'float'
                , 'flr_lvl':'int'
                , 'window_n':'int'
                , 'window_type':'category'
                , 'door_type':'category'
                , 'mortgage':'category'
                , 'district':'category'
               })
      )

df3.info()
```

```python
# !!! code needs improving
def clean_area_col(area,name,desc):
    
    if ((np.ceil(area) == np.ceil(name) == np.ceil(desc)) | (np.floor(area) == np.floor(name) == np.floor(desc))):
        return max(area, name, desc)
    
    elif (name == 99999) & (desc == 99999):
        return area
    
    elif name == 99999:
        if (np.ceil(area) == np.ceil(desc)) | (np.floor(area) == np.floor(desc)):
            return max(area, desc)
        elif desc > 500:
            return area
        elif area > 500:
            return desc
        else:
            return max(area,desc)
        
    elif desc == 99999:
        if (np.ceil(area) == np.ceil(name)) | (np.floor(area) == np.floor(name)):
            return max(area, name)
        elif name > 500:
            return area
        elif area > 500:
            return name
        else:
            return max(area,name)
    
    elif area > 500:
        return max(name,desc)
    
    elif name > 500:
        return max(area, desc)
    
    elif desc > 500:
        return max(area,name)
    
    else:
        return max(area, name, desc)
  
```

```python
# extract potential missing values from posting title and description
df4 = (df3
       .assign(area_name = lambda x: x.posting_name.str.extract(r'(\d+,?\.?\d+\s?)[–ºm][¬≤2k–∫]'))
       .assign(area_name = lambda x: x.area_name.str.replace(',', '.'))
       .assign(area_name = lambda x: x.area_name.fillna(99999))
       .assign(area_desc = lambda x: x.description.str.extract(r'(\d+,?\.?\d+\s?)[–ºm][¬≤2k–∫]'))
       .assign(area_desc = lambda x: x.area_desc.str.replace(',', '.'))
       .assign(area_desc = lambda x: x.area_desc.fillna(99999))
       .astype({'area_name':'float',
               'area_desc':'float'})
      )

df4['area_m2'] = df4.apply(lambda x: clean_area_col(x['area'], x['area_name'], x['area_desc']), axis=1)

df4
```

```python
# if price less than 20 then usually indicates price per m2
def adjust_price(price, area):
    if price < 20:
        return price * area
    else:
        return(price)
```

```python
# Prepare for analysis
df5 = (df4
       .drop(columns=['area', 'posting_name', 'date_posted', 'description', 'area_name', 'area_desc', 'posting_link'])
       .query('area_m2 >= 15 & area_m2 < 500')
       .query('price > 0.1')
       .reset_index(drop=True)
      )

df5['price'] = df5.apply(lambda x: adjust_price(x['price'], x['area_m2']), axis=1)

df5.head()
```
